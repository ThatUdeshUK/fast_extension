```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
fast = pd.read_csv("../results_remote/output_places_US_legacy_fast_sr.csv")
ckmbr = pd.read_csv("../results_remote/output_places_US_legacy_ckqst_sr.csv")


res = pd.concat([
    fast,
    ckmbr,
])
res = res[res['name'].isin(['legacy_fast', 'legacy_ckqst'])]

name_map = {
    "legacy_fast": "FAST (MBR)",
    "legacy_ckqst": "AOIQ (MBR)",
}
res['name'] = res['name'].apply(lambda x: name_map[x])
res = res[(res['num_queries'] == 100000) & (res['num_objects'] == 1000000)]

res
```

```python
search_res = res.loc[:, ['creation_time', 'sr_rate', 'name']]
search_res['Index creation time (sec)'] = search_res['creation_time'].apply(lambda x: x) / 1000000000
search_res = search_res.rename(columns={"sr_rate": "Spatial Range"})

sns.lineplot(x="Spatial Range", y='Index creation time (sec)', style='name', data=search_res, markers=True)
```

```python
search_res = res.loc[:, ['search_time', 'sr_rate', 'name']]
search_res['Query Matching Time (nano sec)'] = search_res['search_time'].apply(lambda x: x) / 100000
search_res['sr_rate'] = search_res['sr_rate'].apply(lambda x: x * 512)
search_res = search_res.rename(columns={"sr_rate": "Query Spatial Range"})
# create_res['experiment'] = create_res['name'].apply(expName)

sns.lineplot(x="Query Spatial Range", y='Query Matching Time (nano sec)', hue='name', style='name', data=search_res, errorbar="sd", markers=True)
```
