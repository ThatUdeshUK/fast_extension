# KNN Descending params

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

## Import data

```python
res_deg = pd.read_csv("../results_remote/fast_knn_params_deg.csv")
res_deg.head()
```

```python
res_ar = pd.read_csv("../results_remote/fast_knn_params.csv")
```

## Plot

```python
df = pd.DataFrame(res_deg) 
df['search_time'] = df['search_time'].apply(lambda x: x / 1000000)
df['search_time'] = df['search_time'].astype(int)
df = df[~(df['knn_deg_ratio'] == 350)]
df
```

```python
search = pd.DataFrame(df.loc[:, ['knn_deg_ratio', 'search_time']]).groupby(['knn_deg_ratio']).mean().reset_index()
search
```

```python
search.to_csv('out_prama.dat', header=False, index=False, sep='\t')
```

```python

# df['knn_ar_thresh'].drop_duplicates()
plt.figure(figsize=(4, 3))
ax = sns.lineplot(x='knn_deg_ratio', y='search_time', style=True, markers=True, errorbar=None, data=df)
ax.get_legend().remove()
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.xlabel("KNN Degradation Threshold")
plt.ylabel("Object Matching Time (milli sec)")
```

```python
df = pd.DataFrame(res_ar)
df['search_time'] = df['search_time'].apply(lambda x: x / 1000000)

sns.lineplot(x='knn_ar_thresh', y='search_time', data=df)
```

```python
df = pd.DataFrame(res_ar)
df = df[df['knn_ar_thresh'] <=30]
df['search_time'] = df['search_time'].apply(lambda x: x / 100000)
df = df[~(df['knn_ar_thresh'] == 3)]
df = df[~(df['knn_ar_thresh'] == 4)]
```

```python
search = pd.DataFrame(df.loc[:, ['knn_ar_thresh', 'search_time']]).groupby(['knn_ar_thresh']).mean().reset_index()
search
```

```python
search.to_csv('out_prama_ar.dat', header=False, index=False, sep='\t')
```

```python

plt.figure(figsize=(4, 3))
ax = sns.lineplot(x='knn_ar_thresh', y='search_time', style=True, markers=True, errorbar=None, data=df)
ax.get_legend().remove()
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.xlabel("Answer Region (AR) Threshold")
plt.ylabel("Object Matching Time (milli sec)")
```
