# KNN Descending params

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

## Import data

```python
res_deg_1 = pd.read_csv("../results_final/FAST_KNN_NO_ObjIndex_param_deg_1.csv")
res_deg_2 = pd.read_csv("../results_final/FAST_KNN_NO_ObjIndex_param_deg_2.csv")
res_deg_3 = pd.read_csv("../results_final/FAST_KNN_NO_ObjIndex_param_deg_3.csv")
res_deg_4 = pd.read_csv("../results_final/FAST_KNN_NO_ObjIndex_param_deg_4.csv")
res_deg_5 = pd.read_csv("../results_final/FAST_KNN_NO_ObjIndex_param_deg_5.csv")

res_deg = pd.concat([res_deg_1, res_deg_2, res_deg_3, res_deg_4, res_deg_5])
res_deg = res_deg[res_deg['creation_time'] != 0]
res_deg.head()
```

```python
# res_ar = pd.read_csv("../results_remote/fast_knn_params.csv")
```

## Plot

```python
df = pd.DataFrame(res_deg) 
df['search_time'] = df['search_time'].apply(lambda x: x / 100000)
df['search_time'] = df['search_time'].astype(int)

df['creation_time'] = df['creation_time'].apply(lambda x: x / 1000000)
df['creation_time'] = df['creation_time'].astype(int)

df = df[~(df['knn_deg_ratio'] == 350)]
df
```

```python
s_df = df.loc[:, ['knn_deg_ratio', 'knn_ar_thresh', 'create_query_idx_mem']]
search = pd.DataFrame(s_df).groupby(['knn_deg_ratio', 'knn_ar_thresh']).mean().reset_index()

plt.figure(figsize=(4, 3))
ax = sns.lineplot(x='knn_deg_ratio', y='create_query_idx_mem', hue='knn_ar_thresh', style=True, markers=True, errorbar=None, data=search)
ax.get_legend().remove()
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.xlabel("KNN Degradation Threshold")
plt.ylabel("Memory")
```

```python
s_df = df[df['knn_ar_thresh'] == 5.0].loc[:, ['knn_deg_ratio', 'search_query_idx_mem', 'search_query_mem']]
s_df['memory'] = (s_df['search_query_idx_mem'] - s_df['search_query_mem']) / (1024 * 1024)

search = pd.DataFrame(s_df.drop(columns=['search_query_idx_mem', 'search_query_mem'])).groupby(['knn_deg_ratio']).mean().reset_index()

# df['knn_ar_thresh'].drop_duplicates()
plt.figure(figsize=(4, 3))
ax = sns.lineplot(x='knn_deg_ratio', y='memory', style=True, markers=True, errorbar=None, data=search)
ax.get_legend().remove()
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.xlabel("KNN Degradation Threshold")
plt.ylabel("Memory")
```

```python
search.to_csv('out_prama.dat', header=False, index=False, sep='\t')
```
