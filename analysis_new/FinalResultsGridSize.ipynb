```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

font = {'size'   : 12}

matplotlib.rc('font', **font)
```

```python
data_dir = "/homes/ukumaras/Projects/fast"

df = pd.read_csv(f"{data_dir}/results_final/FAST_KNN_NO_ObjIndex_twitter_grid.csv")

df = pd.DataFrame(df).groupby(['name', 'num_queries']).mean().reset_index() #[df.num_queries == 5000000]
df['index'] = df['name'].apply(lambda x: x.split("_")[0])

df['dataset'] = df['name'].apply(lambda x: x.split("_")[-2])
df['grid_size'] = df['name'].apply(lambda x: x.split("_")[-1]).astype(int)
df['num_queries'] = (df['num_queries'] / 1000000).astype(int)
df
```

### Create

```python
create = pd.DataFrame(df.loc[:, ['index', 'grid_size', 'creation_time']])
create = create.sort_values('grid_size')
create
```

```python
out = create.pivot(index='grid_size', columns='index', values='creation_time').reset_index()
out.to_csv('out_grid.dat', header=False, index=False, sep='\t')
```

### Search

```python
search = pd.DataFrame(df.loc[:, ['index', 'grid_size', 'search_time']])
search = search.sort_values('grid_size')
search
```

```python
out = search.pivot(index='grid_size', columns='index', values='search_time').reset_index()
out.to_csv('out_grid.dat', header=False, index=False, sep='\t')
```
