```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

font = {'size'   : 12}

matplotlib.rc('font', **font)
```

```python
data_dir = "/homes/ukumaras/Projects/fast"

twitter_base = pd.read_csv(f"{data_dir}/results_remote_new/CkQST_KNN_NO_twitter_maxk3_ckqst.csv")
twitter_adopt = pd.read_csv(f"{data_dir}/results_final/AdoptCkQST_KNN_NO_twitter_adoptckqst.csv")
twitter_fast = pd.read_csv(f"{data_dir}/results_remote_new/FAST_KNN_NO_ObjIndex_twitter_maxk3_fast.csv")

df = pd.concat([twitter_base, twitter_adopt, twitter_fast])
df = pd.DataFrame(df).groupby(['name', 'num_queries']).mean().reset_index() #[df.num_queries == 5000000]
df['index'] = df['name'].apply(lambda x: x.split("_")[0])

df['dataset'] = df['name'].apply(lambda x: x.split("_")[-2])
df['num_queries'] = (df['num_queries'] / 1000000).astype(int)
df
```

### Create

```python
create = pd.DataFrame(df.loc[:, ['index', 'num_queries', 'creation_time']])
create
```

```python
out = create.pivot(index='num_queries', columns='index', values='creation_time').reset_index()
out.to_csv('out_numq.dat', header=False, index=False, sep='\t')
```
